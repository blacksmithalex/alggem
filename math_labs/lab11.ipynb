## Упражнение C1

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, sin, cos, exp, solve, limit, oo

# Задача 1: Точки перегиба для f(x) = sin(x) - x^2 * cos(x)

# Определение функции
x = symbols('x')
f1 = sin(x) - x**2 * cos(x)

# Построение графика
x_vals_1 = np.linspace(-10, 10, 1000)
f_vals_1 = [f1.subs(x, val) for val in x_vals_1]

plt.figure(figsize=(10, 6))
plt.plot(x_vals_1, f_vals_1, label=r"$f(x) = \sin(x) - x^2 \cos(x)$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("График функции и точки перегиба")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## Упражнение C2

```python
# Задача 2: Исследование функции f(x) = e^(1 / (x^2 - 1))

# Определение функции
f2 = exp(1 / (x**2 - 1))

# Первая производная
f2_prime = diff(f2, x)

# Вторая производная
f2_double_prime = diff(f2_prime, x)

# Нахождение точек экстремума (решение f'(x) = 0)
extrema_points = solve(f2_prime, x)
extrema_points = [p.evalf() for p in extrema_points if p.is_real]

# Нахождение точек перегиба (решение f''(x) = 0)
inflection_points_2 = solve(f2_double_prime, x)
inflection_points_2 = [p.evalf() for p in inflection_points_2 if p.is_real]

# Односторонние пределы в точках разрыва
limits_at_1_plus = limit(f2, x, 1, dir='+')
limits_at_1_minus = limit(f2, x, 1, dir='-')
limits_at_minus_1_plus = limit(f2, x, -1, dir='+')
limits_at_minus_1_minus = limit(f2, x, -1, dir='-')

# Построение графика
x_vals_2 = np.linspace(-2, 2, 1000)
x_vals_2 = x_vals_2[np.abs(x_vals_2) != 1]  # Исключаем точки разрыва
f_vals_2 = [f2.subs(x, val) for val in x_vals_2]

plt.figure(figsize=(12, 8))
plt.plot(x_vals_2, f_vals_2, label=r"$f(x) = e^{1 / (x^2 - 1)}$", color="blue")

# Обозначение точек экстремума и перегиба
plt.scatter(extrema_points, [f2.subs(x, p) for p in extrema_points],
            color="green", label="Экстремумы", zorder=5)
plt.scatter(inflection_points_2, [f2.subs(x, p) for p in inflection_points_2],
            color="red", label="Точки перегиба", zorder=5)

# Обозначение точек разрыва
plt.axvline(1, color="orange", linestyle="--", label="Разрыв в x=1")
plt.axvline(-1, color="orange", linestyle="--", label="Разрыв в x=-1")

plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("График функции, экстремумы и точки перегиба")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

print("Экстремумы для f(x) = e^(1 / (x^2 - 1)):", extrema_points)
print("Точки перегиба для f(x) = e^(1 / (x^2 - 1)):", inflection_points_2)
print("Односторонние пределы:")
print("lim x->1+:", limits_at_1_plus)
print("lim x->1-:", limits_at_1_minus)
print("lim x->-1+:", limits_at_minus_1_plus)
print("lim x->-1-:", limits_at_minus_1_minus)
```
