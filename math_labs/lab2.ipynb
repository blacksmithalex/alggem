## Упражнение С1:

```python
import numpy as np
import matplotlib.pyplot as plt

# Задание 1
# График y = sqrt(x + 3)
x1 = np.linspace(-3, 3, 100)
y1 = np.sqrt(x1 + 3)

# График y = sign(x)
x2 = np.linspace(-5, 5, 100)
y2 = np.sign(x2)

plt.figure(figsize=(12, 6))

# Подзадача (а): y = sqrt(x + 3)
plt.subplot(1, 2, 1)
plt.plot(x1, y1, color='blue', linestyle='--', marker='o', label=r'$y = \sqrt{x + 3}$')
plt.title("График y = sqrt(x + 3)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Подзадача (б): y = sign(x)
plt.subplot(1, 2, 2)
plt.plot(x2, y2, color='red', linestyle='-', marker='s', label=r'$y = \mathrm{sign}(x)$')
plt.title("График y = sign(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## Упражнение С2:

```python
# Задание 2
# Графики y = e^x, y = ln(x), y = x
x_exp = np.linspace(-2, 2, 100)
y_exp = np.exp(x_exp)

x_ln = np.linspace(np.exp(-2), np.exp(2), 100)
y_ln = np.log(x_ln)

x_line = np.linspace(-2, np.exp(2), 100)
y_line = x_line

plt.figure(figsize=(10, 6))
plt.plot(x_exp, y_exp, color='green', linestyle='--', marker='o', label=r'$y = e^x$')
plt.plot(x_ln, y_ln, color='blue', linestyle='-', marker='s', label=r'$y = \ln(x)$')
plt.plot(x_line, y_line, color='red', linestyle='-.', marker='^', label=r'$y = x$')

plt.title("Графики y = e^x, y = ln(x), y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

## Упражнение С3:

```python
# Задание 3
# y = x, y = 1/x, y = 1/sqrt(x)
x3 = np.linspace(0.1, 5, 100)
x4 = np.linspace(0.1, 5, 100)

y_x = x4
y_1_x = 1 / x4
y_1_sqrt_x = 1 / np.sqrt(x4)

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Линейная шкала
axs[0].plot(x3, y_x, color='blue', linestyle='-', label=r'$y = x$')
axs[0].plot(x3, y_1_x, color='green', linestyle='--', label=r'$y = \frac{1}{x}$')
axs[0].plot(x3, y_1_sqrt_x, color='red', linestyle='-.', label=r'$y = \frac{1}{\sqrt{x}}$')
axs[0].set_title("Линейная шкала")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].grid(True)
axs[0].legend()

# Логарифмическая шкала
axs[1].plot(x3, y_x, color='blue', linestyle='-', label=r'$y = x$')
axs[1].plot(x3, y_1_x, color='green', linestyle='--', label=r'$y = \frac{1}{x}$')
axs[1].plot(x3, y_1_sqrt_x, color='red', linestyle='-.', label=r'$y = \frac{1}{\sqrt{x}}$')
axs[1].set_title("Логарифмическая шкала")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].set_xscale('log')
axs[1].set_yscale('log')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

## Упражнение С4:

```python
# Задание 4
# График y = sin(x) и обратная функция
import warnings
warnings.filterwarnings('ignore')
x_sin = np.linspace(-np.pi/2, np.pi/2, 100)
y_sin = np.sin(x_sin)
y_inverse = np.arcsin(x_sin)

plt.figure(figsize=(8, 6))
plt.plot(x_sin, y_sin, color='blue', linestyle='-', label=r'$y = \sin(x)$')
plt.plot(x_sin, y_inverse, color='red', linestyle='--', label=r'$y = \arcsin(x)$')
plt.plot(x_sin, x_sin, color='purple', linestyle='-', label="y = x (ось симметрии)")

plt.title("Графики y = sin(x) и y = arcsin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
