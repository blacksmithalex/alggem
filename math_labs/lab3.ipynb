## Упражнение C1:

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Вычисления с мнимой единицей
i = 1j  # мнимая единица

# 1.1 (1 + 2i)^2 * (1 - i)
result_1_1 = (1 + 2 * i)**2 * (1 - i)

# 1.2 (1 - i)^4 - (1 + i)^4
result_1_2 = (1 - i)**4 - (1 + i)**4

# 1.3 i^k, где k = 20, 21, ..., 40
k = np.arange(20, 41)  # массив степеней от 20 до 40
result_1_3 = i**k

# 1.4 (2 + 5i) / (1 - 3i) + 1 / (1 - i)
result_1_4 = (2 + 5 * i) / (1 - 3 * i) + 1 / (1 - i)

# Вывод результатов для пункта 1
print("1.1:", result_1_1)
print("1.2:", result_1_2)
print("1.3:", result_1_3)
print("1.4 (в разных форматах):")
print("   Комплексное:", result_1_4)
print("   Действительная часть:", result_1_4.real)
print("   Мнимая часть:", result_1_4.imag)
print("   Модуль:", abs(result_1_4))
print("   Аргумент:", np.angle(result_1_4))
```

## Упражнение C2:

```python
# 2. Операции с сопряжёнными числами
z1 = np.sqrt(3) - i
z2 = 1 + i * np.sqrt(3)

# 2.1 z1 * conjugate(z2)
result_2_1 = z1 * np.conj(z2)

# 2.2 (conjugate(z1) / z2)^2
result_2_2 = (np.conj(z1) / z2)**2

print("\n2.1:", result_2_1)
print("2.2:", result_2_2)
```

## Упражнение C3:

```python
# 3. Модуль, аргумент и сопряжённое число
# 3.1 (5 + 4i)(3 - 2i^3)
result_3_1 = (5 + 4 * i) * (3 - 2 * i**3)

# 3.2 (1 - i)^13
result_3_2 = (1 - i)**13

print("\n3.1:", result_3_1)
print("   Модуль:", abs(result_3_1))
print("   Аргумент:", np.angle(result_3_1))
print("   Сопряжённое:", np.conj(result_3_1))

print("3.2:", result_3_2)
print("   Модуль:", abs(result_3_2))
print("   Аргумент:", np.angle(result_3_2))
print("   Сопряжённое:", np.conj(result_3_2))
```

## Упражнение C4:

```python
# 4. Построение корней 6-й степени числа (-1 - sqrt(3)i)
complex_number = -1 - np.sqrt(3) * i
roots = [complex_number**(1/6) * np.exp(2j * np.pi * n / 6) for n in range(6)]

# Параметры окружности
radius = abs(complex_number)**(1/6)
angles = np.linspace(0, 2 * np.pi, 100)

# Построение
plt.figure(figsize=(8, 8))
plt.plot(radius * np.cos(angles), radius * np.sin(angles), 'k--', label="Окружность")

# Добавляем точки и многоугольник
roots_x = [root.real for root in roots]
roots_y = [root.imag for root in roots]
plt.scatter(roots_x, roots_y, color='red', label="Корни", zorder=5)
plt.plot(roots_x + [roots_x[0]], roots_y + [roots_y[0]], 'b-.', label="Многоугольник")

plt.axhline(0, color='black', linewidth=1.2)
plt.axvline(0, color='black', linewidth=1.2)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.title("Корни 6-й степени числа (-1 - sqrt(3)i)")
plt.legend()
plt.show()
```

## Упражнение C5:

```python
# 5. Построение множества точек с заданным аргументом
# Arg(z) = -3pi/4
theta1 = -3 * np.pi / 4
r1 = np.linspace(0, 10, 100)  # радиусы для построения
x1 = r1 * np.cos(theta1)
y1 = r1 * np.sin(theta1)

# Arg(z) = -pi/3
theta2 = -np.pi / 3
r2 = np.linspace(0, 10, 100)
x2 = r2 * np.cos(theta2)
y2 = r2 * np.sin(theta2)

plt.figure(figsize=(8, 8))
plt.plot(x1, y1, 'r-', label=r'Arg(z) = $-\frac{3\pi}{4}$')
plt.plot(x2, y2, 'b-', label=r'Arg(z) = $-\frac{\pi}{3}$')
plt.axhline(0, color='black', linewidth=1.2)
plt.axvline(0, color='black', linewidth=1.2)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.title("Множество точек с заданным аргументом")
plt.legend()
plt.show()
```
