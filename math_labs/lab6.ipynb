## Упражнение С1

```python
import numpy as np
import matplotlib.pyplot as plt

# Задача 1: Построение графиков функций

# Определение функций
def func1(x):
    return np.abs((2 * x - 1) / (x - 1))

def func2(x):
    return np.abs(3 * x - 6) / (x - 2)

# Промежутки для построения графиков
x1 = np.linspace(-10, 10, 1000)
x1 = x1[x1 != 1]  # Исключаем x = 1 (разрыв)
x2 = np.linspace(-10, 10, 1000)
x2 = x2[x2 != 2]  # Исключаем x = 2 (разрыв)

# График для первой функции
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x1, func1(x1), label=r"$y = \left|\frac{2x - 1}{x - 1}\right|$", color='blue')
plt.axvline(1, color='red', linestyle='--', label="x = 1 (разрыв)")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("График функции 1")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(0, 10)
plt.grid(True)
plt.legend()

# График для второй функции
plt.subplot(1, 2, 2)
plt.plot(x2, func2(x2), label=r"$y = \frac{|3x - 6|}{x - 2}$", color='green')
plt.axvline(2, color='red', linestyle='--', label="x = 2 (разрыв)")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("График функции 2")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(0, 10)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## Упражнение С2

```python
# Задача 2: Исследование пределов

# 2a) График ln(x) * sin(x)
x3 = np.linspace(1, 100, 1000)
y3 = np.log(x3) * np.sin(x3)

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x3, y3, label=r"$y = \ln(x) \cdot \sin(x)$", color='purple')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title(r"График $y = \ln(x) \cdot \sin(x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# 2b) График последовательности 2^n / n!
y4 = [2**n / np.math.factorial(n) for n in range(1, 50)]

plt.subplot(2, 1, 2)
plt.plot(n, y4, label=r"$y = \frac{2^n}{n!}$", marker='o', color='orange')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title(r"График $y = \frac{2^n}{n!}$")
plt.xlabel("n")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
