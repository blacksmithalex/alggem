## Упражнение C1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Задача 1: Исследование точек разрыва

# a) y = arcsin(x) / x
def func_a(x):
    return np.arcsin(x) / x

x_a = np.linspace(-1, 1, 500)
x_a = x_a[x_a != 0]  # Исключаем x = 0, где функция может быть неопределена

# b) y = 2 / (x^2 - 1)
def func_b(x):
    return 2 / (x**2 - 1)

x_b = np.linspace(-2, 2, 500)
x_b = x_b[(x_b != -1) & (x_b != 1)]  # Исключаем x = ±1, где функция разрывна

# Графики
plt.figure(figsize=(12, 6))

# График (а)
plt.subplot(1, 2, 1)
plt.plot(x_a, func_a(x_a), label=r"$y = \frac{\arcsin(x)}{x}$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="red", linestyle="--", label="x = 0 (возможный разрыв)")
plt.title("Функция (а)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# График (б)
plt.subplot(1, 2, 2)
plt.plot(x_b, func_b(x_b), label=r"$y = \frac{2}{x^2 - 1}$", color="green")
plt.axvline(-1, color="red", linestyle="--", label="x = -1 (разрыв)")
plt.axvline(1, color="red", linestyle="--", label="x = 1 (разрыв)")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("Функция (б)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## Упражнение C2

```python
import warnings
warnings.filterwarnings('ignore')

# Задача 2: Корни уравнения x^2 - 5x * sin(3x) + 3 = 0

# Определим функцию
def equation_2(x):
    return x**2 - 5 * x * np.sin(3 * x) + 3

# Поиск корней
x_initial_guesses = np.linspace(-10, 10, 20)  # Начальные приближения
roots = []
for guess in x_initial_guesses:
    root = fsolve(equation_2, guess)[0]
    if not any(np.isclose(root, r, atol=1e-5) for r in roots):  # Уникальные корни
        roots.append(root)

# График
x_vals = np.linspace(-10, 10, 1000)
y_vals = equation_2(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$y = x^2 - 5x \cdot \sin(3x) + 3$", color="purple")
plt.axhline(0, color="black", linestyle="--")
plt.scatter(roots, np.zeros_like(roots), color="red", label="Корни", zorder=5)
plt.title("График функции и её корни")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

print("Корни уравнения x^2 - 5x * sin(3x) + 3 = 0:", roots)
```

## Упражнение C3

```python
# Задача 3: Решение уравнений

# 3a) x + 1/x = 2
def equation_3a(x):
    return x + 1 / x - 2

x_initial_guesses_a = [0.1, 10]  # Начальные приближения
roots_3a = [fsolve(equation_3a, guess)[0] for guess in x_initial_guesses_a]

# График
x_vals_a = np.linspace(0.1, 10, 1000)
y_vals_a = x_vals_a + 1 / x_vals_a

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x_vals_a, y_vals_a, label=r"$y = x + \frac{1}{x}$", color="orange")
plt.axhline(2, color="black", linestyle="--", label="y = 2")
plt.scatter(roots_3a, [2, 2], color="red", label="Корни", zorder=5)
plt.title("Уравнение (а)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# 3b) ln(1 + x) = -ln(1 - x^2)
def equation_3b(x):
    return np.log(1 + x) + np.log(1 - x**2)

x_initial_guesses_b = [-0.5, 0.5]  # Начальные приближения
roots_3b = [fsolve(equation_3b, guess)[0] for guess in x_initial_guesses_b]

# График
x_vals_b = np.linspace(-0.99, 0.99, 1000)
y_vals_ln1 = np.log(1 + x_vals_b)
y_vals_ln2 = -np.log(1 - x_vals_b**2)

plt.subplot(2, 1, 2)
plt.plot(x_vals_b, y_vals_ln1, label=r"$y = \ln(1 + x)$", color="blue")
plt.plot(x_vals_b, y_vals_ln2, label=r"$y = -\ln(1 - x^2)$", color="green")
plt.scatter(roots_3b, [np.log(1 + root) for root in roots_3b], color="red", label="Корни", zorder=5)
plt.title("Уравнение (б)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

print("Корни уравнения x + 1/x = 2:", roots_3a)
print("Корни уравнения ln(1 + x) = -ln(1 - x^2):", roots_3b)
```
