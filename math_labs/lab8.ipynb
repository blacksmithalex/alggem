## Упражнение 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import symbols, sin, cos, diff, log, acos, pi

# Задача 1: Отношение приращения функции к приращению аргумента
def delta_f(f, x, delta_x):
    """Вычисляет отношение приращения функции к приращению аргумента."""
    return (f(x + delta_x) - f(x)) / delta_x

# Функция y = (sin(1/x))^5
def func_y(x):
    if x == 0:
        return np.nan  # Исключаем деление на 0
    return (np.sin(1 / x))**5

# Точки и приращения
x_values = [1, 2, -3]
delta_x_values = [0.001, -0.001]

# Вычисляем значения отношения приращения
results_1 = {x: [delta_f(func_y, x, dx) for dx in delta_x_values] for x in x_values}
print(results_1)
```

## Упражнение 2

```python
# Задача 2: Производная функции и таблица результатов
# Вычисление аналитической производной
x_sym = symbols('x')
y_sym = sin(1 / x_sym)**5
dy_dx_sym = diff(y_sym, x_sym)

# Вычисление производной в точках
derivative_values = [float(dy_dx_sym.subs(x_sym, x)) for x in x_values]

# Создание таблицы
table_data = {
    "Точка": x_values,
    "Приращение Δx=0.001": [results_1[x][0] for x in x_values],
    "Приращение Δx=-0.001": [results_1[x][1] for x in x_values],
    "Производная": derivative_values
}
table = pd.DataFrame(table_data).set_index("Точка")
table
```

## Упражнение 3

```python
# Задача 3: Производная функции f(x)
# f(x) = log_3(x^2 + 1) / (arccos(sin(x)))^3
f_sym = log(x_sym**2 + 1, 3) / (acos(sin(x_sym))**3)
df_dx_sym = diff(f_sym, x_sym)

# Значение производной в точке x0 = 0.5
x0 = 0.5
df_dx_value = float(df_dx_sym.subs(x_sym, x0))

# Построение графиков
x_vals = np.linspace(0.1, 2, 500)
f_vals = [float(f_sym.subs(x_sym, x)) for x in x_vals]
df_vals = [float(df_dx_sym.subs(x_sym, x)) for x in x_vals]

# Вывод таблицы и графиков
print("Таблица значений (Задача 2):")
print(table)

plt.figure(figsize=(12, 6))

# График функции f(x)
plt.subplot(2, 1, 1)
plt.plot(x_vals, f_vals, label=r"$f(x)$", color="blue")
plt.title(r"График $f(x)$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

# График производной
plt.subplot(2, 1, 2)
plt.plot(x_vals, df_vals, label=r"$f'(x)$", color="green")
plt.scatter([x0], [df_dx_value], color="red", label=f"f'({x0}) ≈ {df_dx_value:.3f}")
plt.title(r"График $f'(x)$")
plt.xlabel("x")
plt.ylabel(r"$f'(x)$")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
