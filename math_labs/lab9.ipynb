## Упражнение С1

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Задача 1: Корни полинома p(x) = x^6 - 1
# Определим полином
p_coeffs = [1, 0, 0, 0, 0, 0, -1]  # Коэффициенты x^6 - 1
p = np.poly1d(p_coeffs)

# Вычислим корни
roots = np.roots(p_coeffs)

# Проверка корней
roots_check = [np.isclose(p(root), 0) for root in roots]

# Построение графика
x_vals = np.linspace(-2, 2, 500)
y_vals = p(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$p(x) = x^6 - 1$", color="blue")
plt.scatter(roots.real, roots.imag, color="red", label="Корни", zorder=5)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("График многочлена и его корни")
plt.xlabel("Re(x)")
plt.ylabel("Im(x)")
plt.grid(True)
plt.legend()
plt.show()

print("Корни полинома p(x) = x^6 - 1:", roots)
print("Проверка корней:", roots_check)
```

## Упражнение С2

```python
# Задача 2: Операции с полиномами
# Определим полиномы
p2 = np.poly1d([1, 1, 1])  # p(x) = x^2 + x + 1
q2 = np.poly1d([1, -1])  # q(x) = x - 1

# Произведение
product = np.polymul(p2, q2)

# Частное и остаток
quotient, remainder = np.polydiv(p2, q2)

# Сумма
sum_poly = np.polyadd(p2, q2)

print("\nПроизведение полиномов:", product)
print("Частное от деления p(x) на q(x):", quotient)
print("Остаток от деления p(x) на q(x):", remainder)
print("Сумма полиномов:", sum_poly)
```

## Упражнение С3

```python
# Задача 3: Ряд Тейлора
# Определяем полином p(x) = x^3 - 2x^2 + 1
x = sp.Symbol('x')
p3 = x**3 - 2*x**2 + 1

# Разложение в ряд Тейлора в точке x = 2
taylor_poly = sp.series(p3, x, x0=2, n=4).removeO()

print("\nРяд Тейлора многочлена p(x) = x^3 - 2x^2 + 1 в точке x = 2:")
print(taylor_poly)
```
