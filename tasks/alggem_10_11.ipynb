## Задание 1.30

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

# Координаты вершин
vertices = np.array([[2, 3, 4], [-5, 1, 0], [2, 7, 1], [-3, 0, 5]])

# Индексы для построения граней
faces = [[vertices[j] for j in [0, 1, 2]],
         [vertices[j] for j in [0, 1, 3]],
         [vertices[j] for j in [0, 2, 3]],
         [vertices[j] for j in [1, 2, 3]]]

# Настройка 3D графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Добавление граней
poly3d = [np.array(face) for face in faces]
ax.add_collection3d(Poly3DCollection(poly3d, facecolors='cyan', linewidths=1, edgecolors='r', alpha=.25))

# Отображение вершин
ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], color='black')

# Подпись точек
for i, txt in enumerate(["A", "B", "C", "D"]):
    ax.text(vertices[i, 0], vertices[i, 1], vertices[i, 2], txt, size=12, color='black')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

## Задание 2.30 

```python
import matplotlib.pyplot as plt
import numpy as np

# Координаты вершин
A = np.array([-2, 1])
B = np.array([1, -5])
C = np.array([-10, -7])

# Построение треугольника
triangle = np.array([A, B, C, A])
plt.plot(triangle[:, 0], triangle[:, 1], label="Треугольник ABC", color="blue")

# Определим уравнение прямой BC: y = kx + b
k_bc = (C[1] - B[1]) / (C[0] - B[0])  # наклон прямой BC
b_bc = B[1] - k_bc * B[0]  # сдвиг по оси y

# Найдём высоту из точки A
# Сначала найдём наклон для перпендикулярной прямой: -1/k_bc
k_height = -1 / k_bc
b_height = A[1] - k_height * A[0]

# Найдём точку пересечения высоты с прямой BC
x_intersect = (b_height - b_bc) / (k_bc - k_height)
y_intersect = k_bc * x_intersect + b_bc
H = np.array([x_intersect, y_intersect])

# Построим высоту
plt.plot([A[0], H[0]], [A[1], H[1]], linestyle="--", color="red", label="Высота из A")

# Построим медиану из точки A
M = (B + C) / 2  # середина стороны BC
plt.plot([A[0], M[0]], [A[1], M[1]], linestyle="--", color="green", label="Медиана из A")

# Построим биссектрису из точки A
AB_dist = np.linalg.norm(B - A)
AC_dist = np.linalg.norm(C - A)
bisector = (AB_dist * C + AC_dist * B) / (AB_dist + AC_dist)
plt.plot([A[0], bisector[0]], [A[1], bisector[1]], linestyle="--", color="purple", label="Биссектриса из A")

# Добавление точек
plt.scatter(*A, color="black")
plt.scatter(*B, color="black")
plt.scatter(*C, color="black")
plt.scatter(*H, color="black")
plt.scatter(*M, color="black")
plt.scatter(*bisector, color="black")

# Подпись точек
plt.text(*A, "A", fontsize=12)
plt.text(*B, "B", fontsize=12)
plt.text(*C, "C", fontsize=12)
plt.text(H[0], H[1], "H", fontsize=12)
plt.text(M[0], M[1], "M", fontsize=12)
plt.text(bisector[0], bisector[1], "Biss", fontsize=12)

plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axis("equal")
plt.show()
```

## Задание 2.45 

```python
import matplotlib.pyplot as plt
import numpy as np

# Определение функции для прямой 5x - 2y - C = 0
def line_y(x, C):
    return (5 * x - C) / 2

x_vals = np.linspace(-10, 10, 400)

# Построение двух параллельных прямых
plt.plot(x_vals, line_y(x_vals, 1), label="5x - 2y - 1 = 0", color="blue")
plt.plot(x_vals, line_y(x_vals, 9), label="5x - 2y - 9 = 0", color="green")

# Расчет расстояния между прямыми
dist = abs(1 - 9) / np.sqrt(5**2 + (-2)**2)
plt.annotate(f"Расстояние ≈ {dist:.2f}", xy=(2, line_y(2, 1)), xytext=(2, line_y(2, 9)), arrowprops=dict(arrowstyle="<->"))

plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

## Задание 3.40

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Определим параметрическое уравнение прямой
t = np.linspace(-10, 10, 100)
x_line = 1 + 3 * t
y_line = 2 + 2 * t
z_line = -3 - 2 * t

# Определим параметры плоскости x - 2y - 2z - 12 = 0
x_plane = np.linspace(-10, 10, 10)
y_plane = np.linspace(-10, 10, 10)
X_plane, Y_plane = np.meshgrid(x_plane, y_plane)
Z_plane = (X_plane - 2 * Y_plane - 12) / 2

# Найдём точку пересечения прямой и плоскости
# Уравнения у нас такие:
# x = 1 + 3t
# y = 2 + 2t
# z = -3 - 2t
# Подставляем их в уравнение плоскости: (1 + 3t) - 2(2 + 2t) - 2(-3 - 2t) = 12
# Решим это уравнение относительно t

from sympy import symbols, Eq, solve

t_sym = symbols('t')
equation = Eq((1 + 3 * t_sym) - 2 * (2 + 2 * t_sym) - 2 * (-3 - 2 * t_sym), 12)
solution = solve(equation, t_sym)[0]

# Найдём координаты точки пересечения
x_intersection = 1 + 3 * solution
y_intersection = 2 + 2 * solution
z_intersection = -3 - 2 * solution

# Визуализация
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Построим прямую
ax.plot(x_line, y_line, z_line, color="blue", label="Прямая")

# Построим плоскость
ax.plot_surface(X_plane, Y_plane, Z_plane, color='cyan', alpha=0.5)

# Отметим точку пересечения
ax.scatter(x_intersection, y_intersection, z_intersection, color="red", s=50, label="Точка пересечения")
ax.text(x_intersection, y_intersection, z_intersection, "Intersection", color='red')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
plt.show()
```

## Задание 4.30

```python
import numpy as np
import matplotlib.pyplot as plt

# Определим диапазон для x и y
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)

# Определим уравнение кривой второго порядка
Z = 3 * X**2 + 3 * X * Y + 6 * X + 4 * Y - 1

# Визуализация
plt.contour(X, Y, Z, levels=[0], colors="blue")
plt.title("Кривая второго порядка: $3x^2 + 3xy + 6x + 4y - 1 = 0$")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axhline(0, color="black",linewidth=0.5)
plt.axvline(0, color="black",linewidth=0.5)
plt.show()
```
