### Задание 1
Лежат ли на одной прямой точки $A(1, -5, 3)$, $B(3, -13, 6)$, $C(-1, 3, 0)$. В случае положительного ответа, написать уравнение этой прямой, проходящей через точку $C$. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Координаты точек
A = np.array([1, -5, 3])
B = np.array([3, -13, 6])
C = np.array([-1, 3, 0])

# Векторы AB и AC
AB = B - A
AC = C - A

# Проверка коллинеарности (AB и AC должны быть пропорциональны)
ratios = AB / AC  # поэлементное деление для нахождения пропорций

# Построение графика для визуализации точек и векторов
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Рисуем точки
ax.scatter(*A, color='r', label='A(1, -5, 3)', s=100)
ax.scatter(*B, color='g', label='B(3, -13, 6)', s=100)
ax.scatter(*C, color='b', label='C(-1, 3, 0)', s=100)

# Рисуем векторы AB и AC
ax.quiver(*A, *AB, color='r', label='AB', arrow_length_ratio=0.1)
ax.quiver(*A, *AC, color='b', label='AC', arrow_length_ratio=0.1)

# Настройки осей
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Точки и векторы AB и AC')

# Добавляем легенду
ax.legend()

plt.show()

# Проверим коллинеарность
ratios
```

### Задание 2
Найти геометрическое место точек, равноудаленных от точек $М(2,1,-2)$ и $N(-2,3,4)$.

```python
# Координаты точек M и N
M = np.array([2, 1, -2])
N = np.array([-2, 3, 4])

# Вектор MN (нормальный вектор к плоскости)
MN = N - M

# Уравнение плоскости: -4x + 2y + 6z - 10 = 0
# Построим плоскость в 3D пространстве

# Создаем сетку значений для x и y
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x_vals, y_vals)

# Выражаем z из уравнения плоскости
z = (-10 + 4*x - 2*y) / 6

# Построение графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Рисуем плоскость
ax.plot_surface(x, y, z, alpha=0.5, rstride=100, cstride=100)

# Отметим точки M и N
ax.scatter(*M, color='r', label='M(2, 1, -2)', s=100)
ax.scatter(*N, color='b', label='N(-2, 3, 4)', s=100)

# Подписи осей
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Геометрическое место точек, равноудаленных от M и N')

# Добавляем легенду
ax.legend()

plt.show()
```

### Задание 3
Найти расстояние от начала координат до плоскости, которая пересекает оси $Ox, Oy, Oz$ в точках с координатами $a = -6, b = 3, c = 3$. 

```python
# Коэффициенты уравнения плоскости
A = -1/6
B = 1/3
C = 1/3
D = -1

# Координаты начала координат (0, 0, 0)
x0, y0, z0 = 0, 0, 0

# Формула расстояния от точки до плоскости
distance = abs(A*x0 + B*y0 + C*z0 + D) / np.sqrt(A**2 + B**2 + C**2)

# Построим график плоскости и визуализируем это расстояние
x_vals = np.linspace(-10, 10, 400)
y_vals = np.linspace(-10, 10, 400)
x, y = np.meshgrid(x_vals, y_vals)

# Выражаем z из уравнения плоскости
z = (1 - A*x - B*y) / C

# Построение графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Рисуем плоскость
ax.plot_surface(x, y, z, alpha=0.5, rstride=100, cstride=100)

# Отметим начало координат и визуализируем расстояние до плоскости
ax.scatter(0, 0, 0, color='r', label='Начало координат (0, 0, 0)', s=100)
ax.quiver(0, 0, 0, A, B, C, length=distance, color='g', label=f'Расстояние = {distance:.2f}', arrow_length_ratio=0.1)

# Подписи осей
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Расстояние от начала координат до плоскости')

# Добавляем легенду
ax.legend()

plt.show()

distance  # Покажем рассчитанное расстояние
```

### Задание 4
Установить взаимное расположение прямой $L$ и плоскости $\alpha$. Прямая $L$ задается параметрически $(2 - 4t, t, -3 + 2t)$, $\alpha$ задается так: $5x - 6y + 2z - 10 = 0$.

```python
A = 5
B = -6
C = 2
D = -10

# Построим график плоскости
x_vals = np.linspace(-10, 10, 400)
y_vals = np.linspace(-10, 10, 400)
x, y = np.meshgrid(x_vals, y_vals)

# Выражаем z из уравнения плоскости
z = (1 - A*x - B*y) / C

# Координаты точек
A = np.array([6, -1, -5])
B = np.array([-2, 1, -1])

# Вектор AB
AB = B - A

# Построение графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Рисуем плоскость
ax.plot_surface(x, y, z, alpha=0.5, rstride=100, cstride=100)

# Рисуем вектор AB (соответсвтующий части нашей прямой)
ax.quiver(*A, *AB, color='r', label='AB', arrow_length_ratio=0.1)

# Точка пересечения
t = - 6 / 22
M = np.array([2 - 4 * t, t, -3 + 2 * t])
ax.scatter(*M, color='r', label='M (точка пересечения)', s=100)

# Подписи осей
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Взаимное расположение плоскости и прямой')

# Добавляем легенду
ax.legend()

plt.show()
```
