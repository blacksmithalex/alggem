## Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Настройки графиков
def plot_settings(title):
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(title)

# 1. Старое начало координат в точке (0, 0), новое начало в (-2, 2), и вектор между ними
def plot_coordinate_shift():
    old_origin = np.array([0, 0])
    new_origin = np.array([-2, 2])

    plt.figure()
    plot_settings("Перенос начала координат")

    # Отметим старое и новое начало координат
    plt.scatter(*old_origin, color='blue', label='Старое начало (0, 0)')
    plt.scatter(*new_origin, color='red', label='Новое начало (-2, 2)')

    # Нарисуем вектор между старыми и новыми координатами
    plt.arrow(old_origin[0], old_origin[1], new_origin[0] - old_origin[0], new_origin[1] - old_origin[1], 
              head_width=0.2, length_includes_head=True, color='green', label='Вектор переноса')

    # Подписи
    plt.text(0.2, -0.2, 'Старое начало', color='blue')
    plt.text(-2.5, 2.2, 'Новое начало', color='red')

    plt.legend()
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.show()

plot_coordinate_shift()
```

## Задание 2

```python
# 2. Точки (sqrt(2), -sqrt(2)) и (0, 2), старый и новый базис
def plot_basis_and_points():
    sqrt2 = np.sqrt(2)
    
    # Точки
    point1 = np.array([sqrt2, -sqrt2])
    point2 = np.array([0, 2])

    # Старый базис
    old_basis_x = np.array([1, 0])
    old_basis_y = np.array([0, 1])

    # Новый базис
    new_basis_x = np.array([-sqrt2/2, sqrt2/2])
    new_basis_y = np.array([-sqrt2/2, -sqrt2/2])

    plt.figure()
    plot_settings("Старый и новый базис, две точки")

    # Отметим точки
    plt.scatter(*point1, color='blue', label=r'Точка ($\sqrt{2}$, $-\sqrt{2}$)')
    plt.scatter(*point2, color='red', label='Точка (0, 2)')

    # Старый базис (синий цвет)
    plt.quiver(0, 0, old_basis_x[0], old_basis_x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Старый базис [1, 0]')
    plt.quiver(0, 0, old_basis_y[0], old_basis_y[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Старый базис [0, 1]')

    # Новый базис (красный цвет)
    plt.quiver(0, 0, new_basis_x[0], new_basis_x[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'Новый базис [$-\frac{\sqrt{2}}{2}$, $\frac{\sqrt{2}}{2}$]')
    plt.quiver(0, 0, new_basis_y[0], new_basis_y[1], angles='xy', scale_units='xy', scale=1, color='red', label=r'Новый базис [$-\frac{\sqrt{2}}{2}$, $-\frac{\sqrt{2}}{2}$]')

    # Подписи
    plt.text(sqrt2 + 0.2, -sqrt2 - 0.2, r'($\sqrt{2}$, $-\sqrt{2}$)', color='blue')
    plt.text(0.2, 2.2, '(0, 2)', color='red')

    plt.legend()
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.show()

plot_basis_and_points()
```

## Задание 3

```python
# 3. Ромб с координатами (-2, 0), (0, -1), (2, 0), (0, 1), сдвиг на произвольное значение
def plot_shifted_rhombus_v2():
    # Координаты вершин ромба
    rhombus = np.array([[-2, 0], [0, -1], [2, 0], [0, 1], [-2, 0]])

    # Произвольный сдвиг
    shift_x, shift_y = 3, 2  # Сдвиг на (3, 2)
    shifted_rhombus = rhombus + np.array([shift_x, shift_y])

    plt.figure()
    plot_settings("Сдвиг ромба (сдвиг на 3, 2)")

    # Исходный ромб
    plt.plot(rhombus[:, 0], rhombus[:, 1], 'b-', label="Исходный ромб", marker='o')
    
    # Сдвинутый ромб
    plt.plot(shifted_rhombus[:, 0], shifted_rhombus[:, 1], 'r--', label="Сдвинутый ромб", marker='o')

    # Подписи
    plt.text(-2.5, 0.2, "(-2, 0)", color='blue')
    plt.text(0.2, -1.2, "(0, -1)", color='blue')
    plt.text(2.5, 0.2, "(2, 0)", color='blue')
    plt.text(0.2, 1.2, "(0, 1)", color='blue')

    plt.text(1, 2.2, "(1, 2)", color='red')
    plt.text(3, 0.8, "(3, 1)", color='red')
    plt.text(5, 2.2, "(5, 2)", color='red')
    plt.text(3, 3.2, "(3, 3)", color='red')

    plt.legend()
    plt.xlim(-4, 6)
    plt.ylim(-3, 5)
    plt.show()

# Выполнение построения
plot_shifted_rhombus_v2()
```

## Задание 4

```python
# 4. Поворот ромба на произвольный угол (например, 45 градусов)
def plot_rotated_rhombus_v2():
    # Координаты вершин ромба
    rhombus = np.array([[-2, 0], [0, -1], [2, 0], [0, 1], [-2, 0]])

    # Поворот на произвольный угол (например, 45 градусов)
    angle = np.radians(45)  # Угол в радианах
    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
    
    rotated_rhombus = np.dot(rhombus, rotation_matrix)

    plt.figure()
    plot_settings("Поворот ромба на 45 градусов")

    # Исходный ромб
    plt.plot(rhombus[:, 0], rhombus[:, 1], 'b-', label="Исходный ромб", marker='o')
    
    # Повернутый ромб
    plt.plot(rotated_rhombus[:, 0], rotated_rhombus[:, 1], 'r--', label="Повернутый ромб", marker='o')

    # Подписи
    plt.text(-2.5, 0.2, "(-2, 0)", color='blue')
    plt.text(0.2, -1.2, "(0, -1)", color='blue')
    plt.text(2.5, 0.2, "(2, 0)", color='blue')
    plt.text(0.2, 1.2, "(0, 1)", color='blue')

    plt.text(1.2, 1.2, "(Повернутая точка)", color='red')

    plt.legend()
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.show()

# Выполнение построения
plot_rotated_rhombus_v2()
```
