## Задание 2

Записать каноническое уравнение кривой второго порядка, определить тип и найти каноническую систему координат. Изобразить кривую $9x^2 - 4xy + 6y^2 + 16x - 8y - 2 = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Уравнение эллипса: 9x^2 - 4xy + 6y^2 + 16x - 8y - 2 = 0
# Преобразуем это уравнение для построения параметрического графика

# Сетка точек для поиска границы эллипса
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Уравнение эллипса
ellipse = 9 * X**2 - 4 * X * Y + 6 * Y**2 + 16 * X - 8 * Y - 2

# Уравнения прямых
x_line = np.linspace(-5, 5, 100)
y_line1 = -x_line / 2  # x + 2y = 0 -> y = -x/2
y_line2 = 2 * x_line + 2  # y - 2x - 2 = 0 -> y = 2x + 2

# Построение графиков
plt.figure(figsize=(8, 8))
plt.contour(X, Y, ellipse, levels=[0], colors='blue', label="Эллипс")  # Эллипс
plt.plot(x_line, y_line1, 'r-', label="x + 2y = 0")  # Прямая 1
plt.plot(x_line, y_line2, 'g-', label="y - 2x - 2 = 0")  # Прямая 2

# Подписи
plt.title("Эллипс и прямые")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.axis('equal')  # Одинаковый масштаб по осям

# Отображение графика
plt.show()
```
