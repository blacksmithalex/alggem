## Задание 2

Записать каноническое уравнение кривой второго порядка, определить тип и найти каноническую систему координат. Изобразить кривую $9x^2 - 4xy + 6y^2 + 16x - 8y - 2 = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Уравнение эллипса: 9x^2 - 4xy + 6y^2 + 16x - 8y - 2 = 0
# Преобразуем это уравнение для построения параметрического графика

# Сетка точек для поиска границы эллипса
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Уравнение эллипса
ellipse = 9 * X**2 - 4 * X * Y + 6 * Y**2 + 16 * X - 8 * Y - 2

# Уравнения прямых
x_line = np.linspace(-5, 5, 100)
y_line1 = -x_line / 2  # x + 2y = 0 -> y = -x/2
y_line2 = 2 * x_line + 2  # y - 2x - 2 = 0 -> y = 2x + 2

# Построение графиков
plt.figure(figsize=(8, 8))
plt.contour(X, Y, ellipse, levels=[0], colors='blue', label="Эллипс")  # Эллипс
plt.plot(x_line, y_line1, 'r-', label="x + 2y = 0")  # Прямая 1
plt.plot(x_line, y_line2, 'g-', label="y - 2x - 2 = 0")  # Прямая 2

# Подписи
plt.title("Эллипс и прямые")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.axis('equal')  # Одинаковый масштаб по осям

# Отображение графика
plt.show()
```

## Задание 3

$9x^2 - 4xy + 6y^2 + 16x - 8y - 2 = 0$

```python
# Импорт необходимых библиотек
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создание сетки координат
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Эллипсоид: 7x^2 + 6y^2 + 5z^2 - 4xy - 4yz - 6x - 24y + 18z + 30 = 0
# Выразим Z для эллипсоида (приблизительное вычисление)
Z_ellipsoid = np.sqrt(
    np.abs(
        (30 - 7 * X**2 - 6 * Y**2 + 4 * X * Y + 6 * X + 24 * Y) / 5
    )
)

# Построение 3D графика с только эллипсоидом
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Эллипсоид
ax.plot_surface(X, Y, Z_ellipsoid, alpha= 0.7, rstride=5, cstride=5, color='blue')
ax.plot_surface(X, Y, -Z_ellipsoid, alpha=0.7, rstride=5, cstride=5, color='blue')  # Нижняя часть эллипсоида

# Подписи осей
ax.set_title("Эллипсоид", fontsize=14)
ax.set_xlabel("X", fontsize=12)
ax.set_ylabel("Y", fontsize=12)
ax.set_zlabel("Z", fontsize=12)

plt.show()
```
